```python
# -*- coding: utf-8 -*-
"""
Created on Fri Jul 17 13:26:58 2020

"""

import pandas as pd
import project_lib as p
from importlib import reload
import matplotlib.pyplot as plt
import numpy as np

 
# %% build and backtest all

reload(p)

def run(asset):
    #load BB data for the asset
    df = p.bbload(asset)

    # create features and signals
    df = p.feature(df)
    s  = p.signal(df)

    #backtest and optimize the x vector. All results in p.F
    p.F['optTarget'] = 'sharpe'
    
    p.backtest()
    p.plotresult()

    
#%% Run all assets and store the PNL, we'll run a optim on it   
 
assets = ['CO1 COMDTY','SPX INDEX','EURUSD CURNCY','CL1 COMDTY','TY1 COMDTY','HG1 COMB COMDTY']
PNL = pd.DataFrame()

for i in range( len(assets) ):
    asset = assets[i]
    print('***************** '+ asset)
    run(asset)
         
    name = p.F['name']
    PNL[name] = p.F['pnl'].copy()

PNL
```

```python
import numpy as np

# %% optimize portfolio
p.PF = PNL
xb = p.pfopt(PNL)

PNLw = PNL*xb  # weighted pnl

pfcumpnl = PNLw.sum(axis=1).cumsum().apply(np.exp)-1

plt.plot(PNL.cumsum().apply(np.exp)-1)
plt.title('Portfolio Components')
plt.show()

v = p.myvars()
n = int(len(pfcumpnl) * v['insamplefactor'])
plt.plot(pfcumpnl)
plt.title('Portfolio Blended OPTIM')
plt.scatter(pfcumpnl.index[n],pfcumpnl[n],color='r')
plt.show()
 
```
