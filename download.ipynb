```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pdblp #don't need that once spreadsheets exist

def bbload(ticker): #don't need that when spreadsheets exist
 
    con = pdblp.BCon(debug=False, port=8194, timeout=9000)
    con.start()
    a=con.bdh(ticker, ['PX_LAST','HIGH','LOW','OPEN'],  '20000101', '20200701' )
   
    return a

```

```python
assets = ['CO1 COMDTY','CO2 COMDTY','TY1 COMB Comdty','SPX INDEX','CCMP INDEX','XAU CURNCY','NG1 COMDTY','NG2 COMDTY','HG1 COMB COMDTY','HG2 COMB COMDTY','EURUSD CURNCY','GBPUSD CURNCY','JPYUSD CURNCY']
names=['co1','co2','ty1','sp500','nasdaq','gold','ng1','ng2','copper1','copper2','eur','gbp','jpy']
mypath = "C:/Users/pandurand/Documents/Algo_Trading/Portfolio_optimization/"
i=0
Prices={}
for asset in assets:
    print(asset)
    d=bbload(asset) #comment that one when spreadsheets exist
    d.to_csv(mypath+asset+".xlsx") #comment that one when spreadsheets exist 
    name=names[i]
    df=pd.read_csv(mypath+asset+".xlsx",index_col='ticker')
    df=df.drop(['field','date'])
    df.columns=['close','high','low','open']
    df.index.name='date'
    Prices[name]=df
    i+=1
```
